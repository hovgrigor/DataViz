## Part 1: Data Cleaning and Exploration

### 1) Load the dataset. Check the first 5 rows

```python
import pandas as pd
import numpy as np
df = pd.read_csv("crime_data.csv")
df.head(5)
```

### 2) Identify columns with missing values and their respective counts. Drop columns where more than 50% of the data is missing (store this version as a new dataset).


```python
missing_values_cols = df.columns[df.isnull().sum() != 0]
counts = {}
for col in list(missing_values_cols):
    counts[col] = df[col].isna().sum()
display(counts)
    
new_df = df.dropna(thresh=len(df.index) // 2, axis=1)  
new_df.head()
```

### 3) Convert the DATE OCC column to a datetime format. Extract the year, month, and day into separate columns. Create a new column for the hour using the TIME OCC column.


```python
new_df["DATE OCC"] = pd.to_datetime(new_df["DATE OCC"])
new_df["Month"] = new_df["DATE OCC"].dt.month
new_df["Year"] = new_df["DATE OCC"].dt.year
new_df["Day"] = new_df["DATE OCC"].dt.day
new_df["Hour"] = new_df["TIME OCC"].apply(lambda x: str(x)[:-2]) #Taking the num before the last two 
new_df["Hour"] = new_df["Hour"].apply(lambda x: int(x) if x else 0) #If the string is empty than it's 12pm
new_df.head()
```

### 4) Filter the dataset for crimes that occurred in 2023. Further filter crimes with the description BURGLARY in the Crm Cd Desc column.

```python
df_2023 = new_df[new_df["Year"] == 2023]
df_bulg_2023 = df_2023[df_2023["Crm Cd Desc"] == "BURGLARY"]
df_bulg_2023.head()
```

### 5) Group the data by AREA NAME and calculate the total number of crimes and the average victim age. Sort the results by total crimes in descending order.

```python
df_bulg_2023.groupby("AREA NAME").agg({"Vict Age": ['count', 'mean']}).sort_values(by=("Vict Age", "count"), ascending=False)
```

## Part 4: Advanced Analysis (Python & R)

### Create a new column, Severity Score, based on the following rules:
• Assign 5 points if a weapon was used (Weapon Used Cd is not null). \
• Assign 3 points for crimes under BURGLARY. \
• Assign 1 point for all other crimes. \
• Group by AREA NAME and find the total severity score for each area.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("crime_data.csv")
```

```python
df["Severity Score"] = df["Weapon Used Cd"].apply(lambda x: 5 if not pd.isna(x) else 0)
df["Severity Score"] += df["Crm Cd Desc"].apply(lambda x: 3 if x == "BURGLARY" else 0)
df["Severity Score"] = df["Severity Score"].apply(lambda x: 1 if x == 0 else x)

df.groupby("AREA NAME")["Severity Score"].sum()
```

## Bonus Part: 

### Use the LAT and LON columns to identify crimes that occurred within a specific latitude-longitude bounding box (e.g., downtown area).

```python
def distance(lat1, lon1, lat2, lon2): #Get distance between two points 
    r = 6371 # km
    p = np.pi / 180

    a = 0.5 - np.cos((lat2-lat1)*p)/2 + np.cos(lat1*p) * np.cos(lat2*p) * (1-np.cos((lon2-lon1)*p))/2
    return 2 * r * np.arcsin(np.sqrt(a))

max_dist = 5 #maximum distance in which two points are considered in the same area

ref_lat = df.iloc[0].LAT
ref_lon = df.iloc[0].LON

df["distance_from_ref"] = distance(ref_lat, ref_lon, df["LAT"], df["LON"])

df[df["distance_from_ref"] < max_dist]
```
